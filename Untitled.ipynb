```python
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
def u(x):
    x=np.array(x)
    L=1.
    u0=1.
    return u0*np.exp(-x**2/L**2)
def BB(xx,t):
    B0=1.
    v0=1.
    L=1.
    Bx=B0*np.ones(np.shape(xx))
    By=-2.*B0*(v0/L**2)*xx*t*np.exp(-xx**2/L**2)
    Bz=np.zeros(np.shape(xx))
    return ([[Bx],[By],[Bz]])
```

```python
fig = plt.figure()
ax = Axes3D(fig)
d=0.2
a=0.4
t=1.
xx = [a,a+d,a+d,a]
yy = [0,0,0,0]
zz = [0,0,d,d]
verts = [zip(xx, yy,zz)]
verts2 = [zip(xx, yy+u(xx),zz)]
verts3 = [zip(xx, yy+u(xx)*2.,zz)]
ax.add_collection3d(Poly3DCollection(verts,alpha=0.2))
ax.add_collection3d(Poly3DCollection(verts2))
ax.add_collection3d(Poly3DCollection(verts3))
ax.quiver3D(XX,YY,ZZ,BB(XX,t)[0],BB(XX,t)[1],BB(XX,t)[2])
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z))

ax.quiver(x, y, z, u, v, w, length=0.1)
```
