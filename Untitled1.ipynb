```python
from mpl_toolkits.mplot3d import axes3d
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
def u(x):
    x=np.array(x)
    L=1.
    u0=1.
    return u0*np.exp(-x**2/L**2)
    
def plot_quiv(t,e=0,a=270):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.view_init(elev=e, azim=a)
    x, y, z = np.meshgrid(np.arange(-2.5, 2.5, 0.25),
                          np.arange(0., 2.5, 0.5),
                          np.arange(0., 2.5, 0.5))
    v0=1.
    L=1.
    B0=1.

    Bx=B0*np.ones(np.shape(x))
    By=-2.*B0*(v0/L**2)*x*t*np.exp(-x**2/L**2)
    Bz=np.zeros(np.shape(z))



    d=0.1
    a=0.5
    xx = [a,a+d,a+d,a]
    yy = [0,0,0,0]
    zz = [0,0,d,d]
    verts = [zip(xx, yy+u(xx)*t,zz)]
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.add_collection3d(Poly3DCollection(verts,alpha=0.6))
    ax.quiver(x, y, z, Bx, By, Bz, length=0.2)
```

```python
plot_quiv(0,e=10,a=270)
```

```python
plot_quiv(1,e=30,a=-40)
```

```python
plot_quiv(3,e=10,a=270)
```

```python
plot_quiv(5,e=10,a=270)
```
