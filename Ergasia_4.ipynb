```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
k=1.38e-23
(rho0,P0,T0)=(1.2,1,300) #kg/m3,atm,K
m=4.8e-26 #kg 
g=10. #m/s2
gamma=5./3
H=k*T0/(m*g) 
```

```python
print H
```

```python
def irho(z):
    return rho0*np.exp(-z/H)
def iP(z):
    return P0*np.exp(-z/H)

def arho(z):
    return rho0*(1-(gamma/(gamma+1))*m*g*z/(k*T0))**(1./(gamma-1))

def aT(z):
    return T0*(1-(gamma/(gamma+1))*m*g*z/(k*T0))
    
def aP(z):
    return P0*(1-(gamma/(gamma+1))*m*g*z/(k*T0))**(1./(gamma-1))
```

```python
z=np.linspace(0,2e4)
```

```python
fig=plt.figure(figsize=(12,16))
ax1=plt.subplot(311)
ax1.plot(z/1000,arho(z),label=u' ')
ax1.plot(z/1000,irho(z),label=u' ')
ax1.set_xlabel(u' (km)')
ax1.set_ylabel(u' $(kg/m^3)$')
ax1.legend()

ax2=plt.subplot(312)
ax2.plot(z/1000,aP(z),label=u' ')
ax2.plot(z/1000,iP(z),label=u' ')
ax2.set_xlabel(u' (km)')
ax2.set_ylabel(u' $(atm)$')
plt.legend()

ax3=plt.subplot(313)
ax3.plot(z/1000,aT(z),label=u' ')
ax3.plot(z/1000,T0*np.ones(z.shape),label=u' ')
ax3.set_xlabel(u' (km)')
ax3.set_ylabel(u' $(K)$')
ax3.set_ylim(-150,350)
ax3.legend(loc='best')

plt.tight_layout()
plt.savefig('Earth_atm',bbox_inches='tight')
```
